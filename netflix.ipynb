**4 Evolution of the catalog: films vs series by year.**

```python
df['year_added'] = df['date_added'].dt.year

types = (
    df.groupby(['year_added', 'type'])
      .size()
      .unstack()
      .fillna(0)
)
```

```python
plt.figure(figsize=(10,6))

plt.plot(types.index,
         types['Movie'],
         label='Movies')

plt.plot(types.index,
         types['TV Show'],
         label='TV Shows')
plt.legend()
```

5 Number of series from 2011 to 2021

```python
df['year'] = df['date_added'].dt.year
```

```python
series = df[df['type'] == 'TV Show']
filtere = series[(series['year'] >= 2011) & (series['year'] <= 2021)]
#add a filter between 2011 and 2021
```

```python
series_by_year = filtere.groupby('year').size()
```

```python
plt.figure(figsize = (10, 6))
sns.countplot(data=filtere, x='year')
```

6 Number of productions per country

```python
top10 = df['country'].value_counts().head(10).index
sns.countplot(data=df[df['country'].isin(top10)], y='country')
```

7 Number of productions by country and age classification

```python
#reduce categories
top10 = df['country'].value_counts().head(10).index
filtered = df[df['country'].isin(top10)]
```

```python
pivot = (
    filtered
    .groupby(['country', 'rating'])
    .size()
    .unstack(fill_value=0)
    #thats unstack function is for ignore multinindex seriess
)
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(pivot)
```

8 Gender analysis (listed_in)

```python
#the firts thing need to do is clean the the columns of listed_in split
df = df.dropna(subset=['listed_in'])

df = df.assign(listed_in=df['listed_in'].str.split(', '))
df = df.explode('listed_in')
```

```python
#count genres
genre_counts = df['listed_in'].value_counts()
top10_genres = df['listed_in'].value_counts().head(10)
```

```python
top10 = df['listed_in'].value_counts().head(10).index
filtered = df[df['listed_in'].isin(top10)]

plt.figure()
sns.countplot(data=filtered, y='listed_in')
```

9 Identify the most frequent content by genre or category.

```python
top10 = genre_counts.head(10)

plt.figure()
sns.barplot(
    x=top10.values,
    y=top10.index
)

plt.title("Top 10 Most Frequent Genres")
```
